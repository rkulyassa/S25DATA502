```python
import pandas as pd

df = pd.read_csv("./data/cereal.csv")
```

2. Create a table with the following manipulations:
- Data selection (columns)
- Data filtering (rows)
- Column renames
- Data transformation (of your choosing)

Make sure to add an effective title, subtitle, and footer. Use a custom font for the heading and
table body.

3. Convert the table above into a heatmap.

```python
# filter columns
list_columns = [
    'name',
    'mfr',
    'rating'
]
df_ratings = df[list_columns]

# filter rows
df_kelloggs_ratings = df_ratings[df_ratings['mfr'] == 'K']

# rename columns
df_kelloggs_ratings = df_kelloggs_ratings.rename(
    columns={
        'name': 'Cereal',
        'mfr': 'Manufacturer',
        'rating': 'Rating'
    }
)

# data transformation: sort descending, and round to tenth
df_kelloggs_ratings = df_kelloggs_ratings.sort_values('Rating', ascending=False)
df_kelloggs_ratings['Rating'] = df_kelloggs_ratings['Rating'].round(1)



# table
from great_tables import (
    GT, md, google_font, style, loc
)

(
    GT(df_kelloggs_ratings[['Cereal', 'Rating']].reset_index(drop=True))
    # add effective title and subtitle
    .tab_header(
        title = md("***All-Bran* is Kellogg's<br>Highest-Rated Cereal**"),
        subtitle = md("Beating out second place<br>*Nutri-grain Wheat* by over 33 points,<br>*All-Bran* secures the top spot in<br>Kellogg's cereal ratings.")
    )
    .opt_align_table_header('left')
    # add footer
    .tab_source_note(source_note=md('**Acknowledgements**<br>Data gathered by Petra Isenberg, Pierre Dragicevic<br>and Yvonne Jansen'))
    .tab_style(
        # custom font for heading
        style=style.text(font=google_font('Lora')),
        locations=loc.title()
    )
    .tab_options(
        heading_title_font_size='26px',
        heading_subtitle_font_size='16px',
        column_labels_font_size='14px',
        table_font_size = '12px',
        data_row_padding = 1,
    )
    # heatmap
    .data_color(
        columns = 'Rating',
        palette = 'Purples',
        domain=[0.0, 100.0]
    )
    # custom font for table body
    .opt_table_font(font=google_font('Sora'))
)
```
