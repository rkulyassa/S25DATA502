1. Download the dataset and load it into a new notebook using the `pd.read_csv()` function.

```python
import pandas as pd

df = pd.read_csv("./data/cereal.csv")
```

2. Recode all the cereal manufacturers to be their full name (e.g. "G" should be replaced with "General Mills", "K" with "Kelloggs", and so forth)

```python
manufacturer_mapping = {
    "N": "Nabisco",
    "Q": "Quaker Oats",
    "K": "Kelloggs",
    "R": "Ralston",
    "G": "General Mills",
    "P": "Post Consumer Brands"
}

df["mfr"] = df["mfr"].replace(manufacturer_mapping)

df
```

3. Create a dataframe called `df_gm` that contains only cereals manufactured by General Mills.

```python
df_gm = df[
    df["mfr"] == "General Mills"
]

df_gm
```

4. Create a dataframe called `df_gm_k` that contains cereals manufactured by either General Mills or Kelloggs.

```python
df_gm_k = df[
    df["mfr"].isin(["General Mills", "Kelloggs"])
]

df_gm_k
```

5. Create a column called `sodium_grams` equal to `sodium` divided by 1000.

```python
df["sodium_grams"] = df["sodium"] / 1000

df
```

6. Rename the `shelf` column as `display_shelf`.

```python
df = df.rename(columns={"shelf": "display_shelf"})

df
```
